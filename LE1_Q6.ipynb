```python
#using python and jupyter notebooks to solve ode first order linear differential equations using separation of variables
#first we have to import our tools
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

#we are gonna make a function that returns dy/dt, in this case we have (y+4)*cos t
def func1 (y, t):
  dydt = (y+4) * np.cos(t)
  return dydt

#we have our initial conditial that is y0 = 0
y0 = 0

#we have to use linspace to disperse our time to plot in our graphic
time = np.linspace(0, 48.8, 1000)

#solvin our ODE
y = odeint(func1, y0, time)

#ploting
plt.plot(time, y, color="red")
plt.xlabel("time")
plt.ylabel("y(t)")
plt.show()
```

```python
#defining the other function
def func2(y, t):
  dydt = y * np.exp(t)
  return dydt

#first value is one
y0 = 1

#using linspace again
time = np.linspace(0, 1.5)

#solving our ODE again
y = odeint(func2, y0, time)

#ploting
plt.plot(time, y, color="purple")
plt.xlabel("time")
plt.ylabel("y(t)")
plt.show()
```
