```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from numpy import cos
from numpy import sin
```

```python
#passing the model that we use to solve our problem
def model(x,t):
  return (x[1],  cos(t) - x[0])
print(model)
y0 = [0, 0] #initial condition
t = np.linspace(0,50,120) #our time range
result = odeint(model,y0,t) #our results

# plot results
plt.plot(t,result[:,0],'purple')
plt.plot(t,result[:,1],'cyan')
plt.xlabel('t')
plt.show()
```

```python
#passing the model that we use to solve our problem manually
t = np.linspace(0,50,120) #our time range
y = t*sin(t)/2

# plot results
plt.plot(t,y,'green')
plt.xlabel('t')
plt.show()
```

## As we can see, the y in our model (x[0], the purple) is the same y in our graphic from the equation we solved manually. So both graphics are correctly.
