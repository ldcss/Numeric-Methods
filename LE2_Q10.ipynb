```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from math import cos
```

```python
#passing the model that we use to solve our problem
def model(x,t):
  return (x[1],  cos(t) - x[0])
y0 = [0, 0] #initial condition
t = np.linspace(0,50,100) #our time range
result = odeint(model,y0,t) #our results

# plot results
plt.plot(t,result[:,0],'purple')
plt.plot(t,result[:,1],'cyan')
plt.xlabel('t')
plt.show()
```
